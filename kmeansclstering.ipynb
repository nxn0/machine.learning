```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('income.csv')
df.head()
```

```python
plt.scatter(df['Age'],df['Income($)'])
```

```python
#importing kmeans from sklearn
from sklearn.cluster import KMeans
```

```python
#creating the model this got a bit difference
model = KMeans(n_clusters=3)
```

```python
y_pred = model.fit_predict(df[['Age','Income($)']])
```

```python
y_pred
```

```python


df.head()
```

```python
df['cluster'] = y_pred
df.head()
```

```python
#dividing DataFrame into 3 different cluster dataframes

df0=df[df.cluster==0]
df1=df[df.cluster==1]
df2=df[df.cluster==2]
```

```python
plt.scatter(df0['Age'],df0['Income($)'], color = 'green' , marker='+')
plt.scatter(df1['Age'],df1['Income($)'], color = 'red' , marker='*')
plt.scatter(df2['Age'],df2['Income($)'], color = 'blue')
plt.xlabel('Age')
plt.ylabel('Income')
```

```python
#this cleary shows that the cluster is wrong hence we do normaliisation,
#all features must be of the same scale
#MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
```

```python
df
```

```python
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])
df
```

```python
#from scatterplot the number of clusters observed is 3
#lets check the same with elbow method

#on x axis its a random k value, y as sse values

k_range = range(1,10)
sse = []
for x in k_range:
    model1 = KMeans(n_clusters=x)
    model1.fit(df[['Age','Income($)']])
    sse.append(model1.inertia_)

```
