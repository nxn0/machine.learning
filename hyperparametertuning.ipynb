```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = iris.data
y = iris.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
#grid search cv

from sklearn.model_selection import GridSearchCV
```

```python
model=GridSearchCV(SVC(),{'C':[1,5,10,15,20,50,100],'kernel':['rbf','linear','poly']},cv =10, return_train_score = False)
```

```python
model.fit(x,y)
```

```python
model.cv_results_
```

```python
df = pd.DataFrame(model.cv_results_)
```
